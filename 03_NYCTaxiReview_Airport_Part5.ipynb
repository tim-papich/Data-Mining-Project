```python
# import data from data files containing samples of the data
sampdf = pd.read_csv('nytAll_1pcnt.csv', dtype='unicode')
sampdfhv = pd.read_csv('hv_combined_sampled_1pcnt.csv', dtype='unicode')
month = 2 # initialize month and year counters that will be used to iterate through the data
year = 2019

# iterate through data files to grab data one month at a time
while (month != 18):
    smpdf = sampdf.loc[(sampdf['year'] == str(year))]
    smpdf = smpdf.loc[(smpdf['month'] == str(month))]
    smpdfhv = sampdfhv.loc[(sampdfhv['year'] == str(year))]
    smpdfhv = smpdfhv.loc[(smpdfhv['month'] == str(month))]
    smpairport()
    if (year != 2020 and month < 12): # increment month if not in the year 2020 and not already at December
        month = month + 1
    elif (month == 12): # if at December 2019 then set to January 2020
        year = year + 1
        month = 1
    elif (month < 6): # go until june 2020
        month = month + 1
    else:
        month = 18 # stop condition
```

```python
# Use average of hv September and November for the missing high volume October data
smphvlag_pick[8]=((smphvlag_pick[7]+smphvlag_pick[9])/2)
smphvjfk_pick[8]=((smphvjfk_pick[7]+smphvjfk_pick[9])/2)
smphvnwk_pick[8]=((smphvnwk_pick[7]+smphvnwk_pick[9])/2)
smphvlag_drop[8]=((smphvlag_drop[7]+smphvlag_drop[9])/2)
smphvjfk_drop[8]=((smphvjfk_drop[7]+smphvjfk_drop[9])/2)
smphvnwk_drop[8]=((smphvnwk_drop[7]+smphvnwk_drop[9])/2)
```

```python
# combine lists of sample data into one dataframe
import statsmodels.formula.api as smf
smpzippedList =  list(zip(lag_psngr, jfk_psngr, nwk_psngr, smplag_pick, smpjfk_pick, smpnwk_pick, smplag_drop, smpjfk_drop, smpnwk_drop, smphvlag_pick, smphvjfk_pick, \
                       smphvnwk_pick, smphvlag_drop, smphvjfk_drop, smphvnwk_drop))
smpnew_df = pd.DataFrame(smpzippedList, columns = ['lag_psngr','jfk_psngr','nwk_psngr','lag_pick','jfk_pick','nwk_pick','lag_drop','jfk_drop','nwk_drop','hvlag_pick', \
                    'hvjfk_pick','hvnwk_pick','hvlag_drop','hvjfk_drop','hvnwk_drop']) 
smpnew_df.head() # view frst five rows of dataframe
```

```python
# build ordinarly least squares regression model for LaGuardia with low p-values and high R-squared
mod = smf.ols('lag_psngr ~ lag_pick + hvlag_pick', data=smpnew_df)
res=mod.fit()
res.summary()
```

```python
# build ordinarly least squares regression model for JFK with low p-values and high R-squared
mod = smf.ols('jfk_psngr ~ jfk_pick', data=smpnew_df)
res=mod.fit()
res.summary()
```

```python
# build ordinarly least squares regression model for Newark with low p-values and high R-squared
mod = smf.ols('nwk_psngr ~ hvnwk_drop', data=smpnew_df)
res=mod.fit()
res.summary()
```
